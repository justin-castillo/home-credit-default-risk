````python
from sklearn.impute import SimpleImputer

zero_impute_cols = [col for col in X.columns if col.startswith(('CC_', 'POS_', 'INSTALL_', 'BUREAU_', 'PREVAPP_'))]
median_impute_cols = [col for col in X.columns if col not in zero_impute_cols]

imputer_zero = SimpleImputer(strategy='constant', fill_value=0)
imputer_median = SimpleImputer(strategy='median')

X_zero = pd.DataFrame(imputer_zero.fit_transform(X[zero_impute_cols]), columns=zero_impute_cols, index=X.index)
X_median = pd.DataFrame(imputer_median.fit_transform(X[median_impute_cols]), columns=median_impute_cols, index=X.index)

X_clean = pd.concat([X_zero, X_median], axis=1)[X.columns]
````

````python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_clean), columns=X_clean.columns, index=X_clean.index)
````

````python
from sklearn.model_selection import train_test_split

X_train_unscaled, X_test_unscaled, y_train, y_test = train_test_split(X_clean, y, test_size=0.2, stratify=y, random_state=42)
````

````python
X_train_scaled = X_scaled.loc[X_train_unscaled.index]
X_test_scaled = X_scaled.loc[X_test_unscaled.index]
````

# 1.3 Modeling

## 1.3.1 Logistic Regression
- Logistic Regression was included as a baseline model due to its simplicity and interpretability. However, its linear assumptions and limited capacity to model complex feature interactions make it a poor fit for high-dimensional structured data like this. As expected, its performance lagged significantly behind gradient boosting methods (e.g., CatBoost, LightGBM), and neither PCA nor hyperparameter tuning yielded meaningful gains. Consequently, it is not considered a serious candidate model for deployment in this context.

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score, classification_report, confusion_matrix

# Baseline Logistic Regression (no grid search, no PCA)
baseline_lr = LogisticRegression(max_iter=1000, class_weight='balanced', solver='liblinear', random_state=42)
baseline_lr.fit(X_train_scaled, y_train)

# Generate predicted probabilities for threshold tuning
y_pred_proba = baseline_lr.predict_proba(X_test_scaled)[:, 1]
````

### Threshold tuning

````python
import numpy as np
import matplotlib.pyplot as plt

y_pred_proba = baseline_lr.predict_proba(X_test_scaled)[:, 1]

thresholds = np.arange(0.05, 0.95, 0.05)

f1s = []
precisions = []
recalls = []

for t in thresholds:
    y_pred = (y_pred_proba > t).astype(int)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    precisions.append(precision)
    recalls.append(recall)
    f1s.append(f1)

    print(f"Threshold: {t:.2f} | Precision: {precision:.3f} | Recall: {recall:.3f} | F1: {f1:.3f}")

# Plot the metrics vs threshold
plt.plot(thresholds, f1s, label='F1 Score')
plt.plot(thresholds, precisions, label='Precision')
plt.plot(thresholds, recalls, label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Logistic Regression Threshold Tuning')
plt.legend()
plt.grid(True)
plt.show()

auc = roc_auc_score(y_test, y_pred_proba)
print(f"Logistic Regression AUC: {auc:.4f}")

y_pred_label = (y_pred_proba > 0.5).astype(int)
print(classification_report(y_test, y_pred_label, digits=3))

coef_df = pd.DataFrame({
    'feature': X_train_scaled.columns,
    'coefficient': baseline_lr.coef_[0]
}).sort_values(by='coefficient', key=abs, ascending=False)

print(coef_df.head(15))
````

````python
y_pred_label = (y_pred_proba > 0.35).astype(int)
print(confusion_matrix(y_test, y_pred_label))
print(classification_report(y_test, y_pred_label, digits=3))
````

## 1.3.2 Random Forest

````python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix 
import matplotlib.pyplot as plt

rfModel1 = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rfModel1.fit(X_train_scaled, y_train)

y_pred_proba_rf = rfModel1.predict_proba(X_test_scaled)[:,1]
y_pred_label_rf = rfModel1.predict(X_test_scaled)

auc_rf = roc_auc_score(y_test, y_pred_proba_rf)
print(f"Random Forest AUC: {auc_rf:.4f}")

print(classification_report(y_test, y_pred_label_rf, digits=3))
print(confusion_matrix(y_test, y_pred_label_rf))

rf_feat_importance = pd.DataFrame({
    'feature': X_train_scaled.columns,
    'importance': rfModel1.feature_importances_
}).sort_values(by='importance', ascending=False)

rf_feat_importance.head(20).plot.bar(x='feature', y='importance', figsize=(10,5), title="Random Forest Feature Importance")
plt.tight_layout()
plt.show()

y_pred_proba_rf = rfModel1.predict_proba(X_test_scaled)[:, 1]
y_pred_rf_thresh = (y_pred_proba_rf > 0.2).astype(int) 

print(classification_report(y_test, y_pred_rf_thresh, digits=3))
print(confusion_matrix(y_test, y_pred_rf_thresh))
````

````python
rfModel2 = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    min_samples_split=10,
    min_samples_leaf=5,
    class_weight='balanced',
    n_jobs=-1,
    random_state=42
)

rfModel2.fit(X_train_scaled, y_train)

y_pred_proba_rf = rfModel2.predict_proba(X_test_scaled)[:,1]
y_pred_label_rf = rfModel2.predict(X_test_scaled)

auc_rf = roc_auc_score(y_test, y_pred_proba_rf)
print(f"Random Forest AUC: {auc_rf:.4f}")

print(classification_report(y_test, y_pred_label_rf, digits=3))
print(confusion_matrix(y_test, y_pred_label_rf))

rf_feat_importance = pd.DataFrame({
    'feature': X_train_scaled.columns,
    'importance': rfModel2.feature_importances_
}).sort_values(by='importance', ascending=False)

rf_feat_importance.head(20).plot.bar(x='feature', y='importance', figsize=(10,5), title="Random Forest Feature Importance")
plt.tight_layout()
plt.show()

y_pred_proba_rf = rfModel2.predict_proba(X_test_scaled)[:, 1]
y_pred_rf_thresh = (y_pred_proba_rf > 0.2).astype(int) 

print(classification_report(y_test, y_pred_rf_thresh, digits=3))
print(confusion_matrix(y_test, y_pred_rf_thresh))
````

````python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

thresholds = np.arange(0.05, 0.95, 0.05)
f1s = []
for t in thresholds:
    preds = (y_pred_proba_rf > t).astype(int)
    f1 = f1_score(y_test, preds)
    f1s.append(f1)
    print(f"Threshold: {t:.2f} | F1: {f1:.3f}")

best_idx = np.argmax(f1s)
optimal_threshold_rf = thresholds[best_idx]
print(f"\nBest threshold = {optimal_threshold_rf:.2f} with F1 = {f1s[best_idx]:.3f}")

y_pred_rf_thresh = (y_pred_proba_rf > optimal_threshold_rf).astype(int)
print(classification_report(y_test, y_pred_rf_thresh, digits=3))
print(confusion_matrix(y_test, y_pred_rf_thresh))
````

### Threshold tuning

- optimal: threshold= 0.60

````python
from sklearn.metrics import f1_score, precision_score, recall_score
import numpy as np
import matplotlib.pyplot as plt

y_pred_proba = cat_model.predict_proba(X_test_unscaled)[:, 1]

thresholds = np.arange(0.05, 0.95, 0.05)

f1s = []
precisions = []
recalls = []

for t in thresholds:
    y_pred = (y_pred_proba > t).astype(int)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    precisions.append(precision)
    recalls.append(recall)
    f1s.append(f1)

    print(f"Threshold: {t:.2f} | Precision: {precision:.3f} | Recall: {recall:.3f} | F1: {f1:.3f}")

# Plot the metrics vs threshold
plt.plot(thresholds, f1s, label='F1 Score')
plt.plot(thresholds, precisions, label='Precision')
plt.plot(thresholds, recalls, label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Threshold Tuning')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import optuna
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score

def objective(trial):
    params = {
        "iterations": 1000,
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
        "depth": trial.suggest_int("depth", 4, 10),
        "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1e-3, 10.0, log=True),
        "bootstrap_type": trial.suggest_categorical("bootstrap_type", ["Bayesian", "Bernoulli", "MVS"]),
        "eval_metric": 'AUC',
        "random_seed": 42,
        "early_stopping_rounds": 50,
        "verbose": False,
        "class_weights": [1, 15], 
    }

    if params["bootstrap_type"] == "Bayesian":
        params["bagging_temperature"] = trial.suggest_float("bagging_temperature", 0, 10)
    elif params["bootstrap_type"] == "Bernoulli":
        params["subsample"] = trial.suggest_float("subsample", 0.5, 1.0)

    model = CatBoostClassifier(**params)

    X_tr, X_val, y_tr, y_val = train_test_split(
        X_train_unscaled, y_train, test_size=0.2, random_state=42, stratify=y_train
    )

    model.fit(
    X_tr, y_tr,
    eval_set=(X_val, y_val),
    use_best_model=True
    )

    preds = model.predict_proba(X_val)[:, 1]
    return roc_auc_score(y_val, preds)
````
